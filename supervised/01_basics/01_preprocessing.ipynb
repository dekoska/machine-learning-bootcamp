```python
import numpy as np
import pandas as pd
import sklearn

sklearn.__version__
```

```python
data = {
    'size': ['XL', 'L', 'M', 'L', 'M'],
    'color': ['red', 'green', 'blue', 'green', 'red'],
    'gender': ['female', 'male', 'male', 'female', 'female'],
    'price': [199.0, 89.0, 99.0, 129.0, 79.0],
    'weight': [500, 450, 300, 380, 410],
    'bought': ['yes', 'no', 'yes', 'no', 'yes']
}

df_raw = pd.DataFrame(data=data)
df_raw
```

```python
df = df_raw.copy()
df.info()
```

```python
for col in ['size', 'color', 'gender', 'bought']:
  df[col] = df[col].astype('category')

df['weight'] = df['weight'].astype('float')
df.info()
```

```python
df.describe() #tylko kolumny numeryczne
```

```python
df.describe().T #transponowanie
```

```python
df.describe(include=['category']).T
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
# le.fit(df['bought'])
# le.transform(df['bought'])
le.fit_transform(df['bought'])
```

```python
 le.classes_
```

```python
df['bought'] = le.fit_transform(df['bought'])
df
```

```python
le.inverse_transform(df['bought'])
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)
encoder.fit(df[['size']])

print(encoder.categories_)
encoder.transform(df[['size']])
```

```python
encoder = OneHotEncoder(drop='first', sparse_output=False)
encoder.fit(df[['size']])
encoder.transform(df[['size']])
```

```python
 pd.get_dummies(data=df, dtype='int')
 pd.get_dummies(data=df, drop_first=True, prefix='new')
 pd.get_dummies(data=df, drop_first=True, prefix_sep='-')
 pd.get_dummies(data=df, drop_first=True, columns=['size'])
 pd.get_dummies(data=df, drop_first=True, dtype='str')
```

odchylenie standardowe - estymator obciony wariancji (numpy), nieobciony (pandas)


```python
print(f"rednia cena: {df['price'].mean()}")
print(f"Odchylenie standardowe ceny: {df['price'].std():.4f}")
```

Standaryzacja:
$ \Huge \frac{x - x.\text{mean()}}{x.\text{std()}}$

```python
def standarize(x):
  return (x-x.mean())/x.std()

standarize(df['price'])
```

```python
from sklearn.preprocessing import scale

scale(df['price'])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['price','weight']] = scaler.fit_transform(df[['price', 'weight']])
df
```
