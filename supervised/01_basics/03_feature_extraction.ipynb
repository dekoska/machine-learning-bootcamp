```python
import pandas as pd
import numpy as np
import sklearn
```

```python
def fetch_financial_data(company='AMZN'):
  import pandas_datareader.data as web
  return web.DataReader(name=company, data_source='stooq')

df_raw = fetch_financial_data()
df_raw.head()
```

```python
df = df_raw.copy()
df = df[:5]
df.info()
```

```python
#generowanie nowych zmiennych
df['day'] = df.index.day
df['month'] = df.index.month
df['year'] = df.index.year
df
```

```python
#dyskretyzacja zmiennej ciaglej
df = pd.DataFrame(data={ 'height' : [175., 178.5, 185, 191, 184.5, 183., 168.]})
df
```

```python
df['height_cut'] = pd.cut(x=df.height, bins=3, labels=['small', 'medium', 'high']) #bins=(160,175,180,195)
df
```

```python
pd.get_dummies(df, drop_first = True, prefix='height', dtype='int')
```

```python
df = pd.DataFrame({'lang' : [['PL', 'ENG'], ['GER', 'ENG', 'PL', 'FRA'], ['RUS']]})
df['lang_numb'] = df['lang'].apply(len)
df['PL_flag'] = df['lang'].apply(lambda x:1 if 'PL' in x else 0)
df
```

```python
df = pd.DataFrame({'website' : ['wp.pl', 'onet.pl', 'google.com']})
new = df.website.str.split('.', expand=True)
df['portal']=new[0]
df['extension']=new[1]
df
```
